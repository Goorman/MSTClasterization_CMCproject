```python
import numpy as np
import pandas as pd
```

```python
z = np.empty(shape = (10, 16))
for i in range(0, 10):
    f = open("outputs/" + str(i+1) + ".txt", "r")
    for k in range(1, 17):
        s = f.readline().strip().split()
        z[i, int(s[1])-1] = float(s[2])
```

```python
for i in range(0, 10):
    for k in range(0, 16):
        d[i, k] = (z[i,0] / z[i,k] ) / (k+1.0) * 100
```

```python
df = pd.DataFrame(data = d)
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(figsize=(10,5))
ax = fig.gca(projection='3d', xlabel = "Number of Threads", ylabel = "Number of Points", zlabel = "Execution Time, sec")
X = np.arange(1, 17, 1)
Y = np.arange(100, 1100, 100)
X, Y = np.meshgrid(X, Y)
Z = z
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm,
                       linewidth=0, antialiased=True)
ax.set_zlim(0, 40)

ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
